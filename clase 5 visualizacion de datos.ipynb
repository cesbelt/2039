```python
path = "datasets/Cartwheeldata.csv"
df = pd.read_csv(path)
df.head(2)
```

```python
sns.lmplot(x = "Wingspan", 
           y = "CWDistance",
          data = df,
          fit_reg = False,
          hue = 'Glasses')
plt.title("Scatter plot of Wingspan vs CWDistance")
plt.show()
```

```python
sns.swarmplot(x='Gender',
            y = 'CWDistance',
            data = df)

plt.title('Gender vs CWDistance')
plt.show()
```

```python
sns.boxplot(data = df.loc[df['Gender'] == 'M',
           ['Height', 'Wingspan', 'CWDistance']]
           )
            
plt.title("box.plot")
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1, 3, 1) # 1 renglon, 2 columns y pos 1
sns.boxplot(data = df.loc[df['Gender'] == 'M',
                         ['Score']
                         ])

ax2 = fig.add_subplot(1, 3, 3) # 1 renglon, 2 columns y pos 2
sns.boxplot(data = df.loc[df['Gender'] == 'F',
                         ['Score']
                         ])
ax1.title.set_text('Male score')
ax2.title.set_text('Female score')

plt.show()
```

```python
sns.distplot(df.CWDistance)

plt.show()
```

```python
sns.countplot(x = 'Gender',
             data = df)
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)

plt.show()
```
